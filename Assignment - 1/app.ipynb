# Q3

```python
def perceptron(x, w, b):
    return 1 if np.dot(x, w) + b >= 0 else 0

X = np.array([[0,0], [0,1], [1,0], [1,1]])

w_and = np.array([1, 1])
b_and = -1.5
print("AND Gate:")
for x in X:
    print(x, perceptron(x, w_and, b_and))

w_or = np.array([1, 1])
b_or = -0.5
print("\nOR Gate:")
for x in X:
    print(x, perceptron(x, w_or, b_or))
```

# Q4

```python
X = torch.tensor([[0.,0.], [0.,1.], [1.,0.], [1.,1.]])
Y = torch.tensor([[0.], [1.], [1.], [0.]])

model = nn.Sequential(
    nn.Linear(2, 4),
    nn.ReLU(),
    nn.Linear(4, 1),
    nn.Sigmoid()
)

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)

for epoch in range(2000):
    optimizer.zero_grad()
    output = model(X)
    loss = criterion(output, Y)
    loss.backward()
    optimizer.step()

print("XOR Output:")
print(model(X).round())
```

# Q5

```python
X = torch.unsqueeze(torch.linspace(0, 10, 100), dim=1)
Y = 2 * X + 1

model = nn.Sequential(
    nn.Linear(1, 10),
    nn.ReLU(),
    nn.Linear(10, 1)
)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(1000):
    optimizer.zero_grad()
    predictions = model(X)
    loss = criterion(predictions, Y)
    loss.backward()
    optimizer.step()

print("Final Loss:", loss.item())
```
